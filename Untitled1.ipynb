```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = yf.download("AAPL", start="2023-07-01", end="2025-07-01")
```

```python
data['SMA10'] = data['Close'].rolling(10).mean()
data['SMA30'] = data['Close'].rolling(30).mean()
```

```python
data['Signal'] = 0
```

```python
for i in range(30, len(data)):
    if data['SMA10'].iloc[i] > data['SMA30'].iloc[i] and data['SMA10'].iloc[i - 1] <= data['SMA30'].iloc[i - 1]:
        data.loc[data.index[i], 'Signal'] = 1 
    elif data['SMA10'].iloc[i] < data['SMA30'].iloc[i] and data['SMA10'].iloc[i - 1] >= data['SMA30'].iloc[i - 1]:
        data.loc[data.index[i], 'Signal'] = -1 
```

```python
data['Position'] = data['Signal'].diff()
```

```python
print("BUY signals (Position == 1):")
buy_signals = data.loc[data['Position'] == 1, ['Close']]
print(buy_signals)

print("\nSELL signals (Position == -1):")
sell_signals = data.loc[data['Position'] == -1, ['Close']]
print(sell_signals)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Close Price', alpha=0.5)
plt.plot(data['SMA10'], label='SMA10', linestyle='--')
plt.plot(data['SMA30'], label='SMA30', linestyle='--')
plt.scatter(buy_signals.index, buy_signals['Close'], label='Buy', marker='^', color='green', s=100)
plt.scatter(sell_signals.index, sell_signals['Close'], label='Sell', marker='v', color='red', s=100)

plt.title("SMA Crossover Strategy: AAPL")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
